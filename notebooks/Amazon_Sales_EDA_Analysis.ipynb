*Features*

* **product_id** - Product ID
* **product_name** - Name of the Product
* **category** - Category of the Product
* **discounted_price** - Discounted Price of the Product
* **actual_price** - Actual Price of the Product
* **discount_percentage** - Percentage of Discount for the Product
* **rating** - Rating of the Product
* **rating_count** - Number of people who voted for the Amazon rating
* **about_product** - Description about the Product
* **user_id** - ID of the user who wrote review for the Product
* **user_name** - Name of the user who wrote review for the Product
* **review_id** - ID of the user review
* **review_title** - Short review
* **review_content** - Long review
* **img_link** - Image Link of the Product
* **product_link** - Official Website Link of the Product



**Step 1  - Import the data**

```python
#Import libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

#Loading the dataset
df = pd.read_csv('/content/sample_data/amazon_sales_data.csv')

#Show the first rows of the dataset
df.head()
```

**Step 2 - Evaluate data quality**

```python
#Verify general information
df.info()
```

```python
#Verify duplicated rows
duplicated_rows= df.duplicated().sum()
print("Duplicated rows: ",duplicated_rows)
```

```python
#Verify null values
df.isnull().sum()
```

**Step 3 - Wrangle the dataset**

```python
#Drop unnecessary columns
df= df.drop(['product_id','about_product','user_id','user_name','review_id','review_title','review_content','img_link','product_link'],axis=1)
df.head()
```

```python
#Creating the first new column 'macro_category'
df['macro_category']= df['category'].str.split('|').str.get(0)
df.head()
```

```python
#Creating the second new column 'micro_category'
df['micro_category']= df['category'].str.split('|').str.get(-1)
df.head()
```

```python
#Drop column 'category'
df= df.drop(['category'],axis=1)
df.head()
```

```python
#Transform the columns 'discounted_price' and 'actual_price' in float values
df['discounted_price']= df['discounted_price'].str.replace('₹','').str.replace(',', '').astype(float)
df['actual_price']= df['actual_price'].str.replace('₹','').str.replace(',', '').astype(float)
df.head()
```

```python
#Transform the column 'discount_percentage' in float values
df['discount_percentage']= df['discount_percentage'].str.replace('%', '').astype(float)
df['discount_percentage']= df['discount_percentage']/100
df.head()
```

```python
#Transform the column 'rating' in float values
df['rating'] = pd.to_numeric(df['rating'].str.replace('|', 'NaN'), errors='coerce')
df.head()
```

```python
#Create column with rate description
df['rate_description']= pd.cut(df['rating'], bins=[1, 2, 3, 4, 5], labels=['Terrible','Poor', 'Average', 'Good'])
df.head()
```

```python
#Transform the column 'rating_count' in float values
df['rating_count']= df['rating_count'].str.replace(',', '').astype(float)
df.head()
```

```python
#Verify null values
df.isnull().sum()
```

```python
#Drop null values
df= df.dropna()
```

```python
#Verify general information
df.info()
```

```python
#Verify statistical information
df.describe()
```

```python
#Show the top 5 most evaluated products in the 'product_name' column
pd.concat([df['product_name'].value_counts(), df['product_name'].value_counts(normalize=True).round(2)], axis=1).head(5)
```

```python
#Show the evaluated products in the 'product_macro_category' column
pd.concat([df['macro_category'].value_counts(), df['macro_category'].value_counts(normalize=True).round(2)], axis=1)
```

```python
#Show the top 5 most evaluated products in the 'micro_category' column
pd.concat([df['micro_category'].value_counts(), df['micro_category'].value_counts(normalize=True).round(2)], axis=1).head(5)
```

```python
#Show the proportion evaluated in the "rate_description" column
pd.concat([df['rate_description'].value_counts(), df['rate_description'].value_counts(normalize=True).round(2)], axis=1)
```

**Step 4 - Evaluete the behavior of the variables**

```python
#Rating histogram
plt.figure(figsize=(8, 4))
sns.histplot(df['rating'], bins=20, kde=True)
plt.title('Rating Histogram')
plt.xlabel('Rating')
plt.ylabel('Frequency')
plt.show()
```

```python
#Numeric correlation matrix
correlation_matrix = df.select_dtypes(include=np.number).corr()['rating']
target_correlation = np.abs(correlation_matrix)
print(target_correlation.sort_values(ascending=False))
```

```python
#Correlation matrix graph
plt.figure(figsize=(10, 6))
numeric_df = df.select_dtypes(include=np.number)
sns.heatmap(numeric_df.corr(), annot=True, cmap='coolwarm')
plt.title('Correlation Matrix')
plt.show()
```

```python
#Correlation line plot off the variables
plt.figure(figsize=(10, 6))
plt.plot(numeric_df.corrwith(numeric_df['rating']).sort_values(ascending=False))
plt.title('Correlation Line Plot')
plt.xlabel('Variables')
plt.ylabel('Correlation')
plt.grid(True)
plt.show()
```

```python
#Scatter plots and histograms
pd.plotting.scatter_matrix(numeric_df, figsize=(10, 10))
plt.show()
```

**Step 5 - Data Analysis**

```python
#Rating Average and Rating Count
sns.lineplot(data=df, x='rating', y='rating_count')
plt.title('Rating Average and Rating Count')
plt.xlabel('Rating Average')
plt.ylabel('Rating Count')
plt.show()
```

```python
#Number of ratings per 'rate_description'
plt.figure(figsize=(10, 6))
ax = sns.countplot(data=df, x='rate_description', palette='RdYlGn')
for p in ax.patches:
    ax.annotate(str(int(p.get_height())),
                (p.get_x() + p.get_width() / 2., p.get_height()),
                ha='center', va='center',
                xytext=(0, 9),
                textcoords='offset points')
plt.title('Number of Ratings per Rate Description')
plt.xlabel('Rate Description')
plt.ylabel('Count')

plt.show()
```

```python
#Most products evalueted per 'macro_category'
most_sold_macro_category = df['macro_category'].value_counts().sort_values(ascending=False)
plt.figure(figsize=(17, 6))
ax = sns.barplot(x=most_sold_macro_category.index, y=most_sold_macro_category.values, palette='viridis')
for p in ax.patches:
    ax.annotate(format(p.get_height(), '.0f'),
                (p.get_x() + p.get_width() / 2., p.get_height()),
                ha='center', va='bottom',
                xytext=(0, 5),
                textcoords='offset points')
plt.title('Most Evalueted Products per Macro Category')
plt.xlabel('Macro Category')
plt.ylabel('Count')
plt.show()
```

```python
#Rating average per "macro_category"
plt.figure(figsize=(17, 6))
ordered_categories = df.groupby('macro_category')['rating'].mean().sort_values(ascending=False).index
ax = sns.barplot(x='macro_category', y='rating', data=df, order=ordered_categories, palette='viridis')
for p in ax.patches:
    ax.annotate(format(p.get_height(), '.1f'),
                (p.get_x() + p.get_width() / 2., p.get_height()),
                ha='center', va='center',
                xytext=(0, 9),
                textcoords='offset points')
plt.title('Rating Average per Macro Category')
plt.xlabel('Macro Category')
plt.ylabel('Rating')
plt.xticks(rotation=45)
plt.show()
```

```python
#Top 10 evalueted products per 'micro_category'
category_order = df['micro_category'].value_counts().nlargest(10).index
ax = sns.countplot(data=df, x='micro_category', palette="viridis", order=category_order)
plt.title('Top 10 evalueted products per Micro Category')
plt.xlabel('Micro Category')
plt.ylabel('Product Count')
plt.xticks(rotation=90)
for p in ax.patches:
    ax.annotate(format(p.get_height(), '.0f'),
                (p.get_x() + p.get_width() / 2., p.get_height()),
                ha='center', va='center',
                xytext=(0, 9),
                textcoords='offset points')

plt.show()
```

```python
#Rating average per 'micro_category' (Top 5 and bottom 5)
plt.figure(figsize=(16, 6))
ordered_micro_categories = df.groupby('micro_category')['rating'].mean().sort_values(ascending=False)
top_3_micro_categories = ordered_micro_categories.head(5)
bottom_3_micro_categories = ordered_micro_categories.tail(5)
selected_micro_categories = pd.concat([top_3_micro_categories, bottom_3_micro_categories])
ax = sns.barplot(x=selected_micro_categories.index, y=selected_micro_categories.values, palette='viridis')
for p in ax.patches:
    ax.annotate(format(p.get_height(), '.1f'),
                (p.get_x() + p.get_width() / 2., p.get_height()),
                ha='center', va='center',
                xytext=(0, 9),
                textcoords='offset points')
plt.title('Rating Average per Micro Category (Top 5 and Bottom 5)')
plt.xlabel('Micro Category')
plt.ylabel('Rating')
plt.show()
```

```python
#Rating average per 'product_name' (top 5 and bottom 5)
plt.figure(figsize=(15, 6))
ordered_products = df.groupby('product_name')['rating'].mean().sort_values(ascending=False)
top_3_products = ordered_products.head(5)
bottom_3_products = ordered_products.tail(5)
selected_products = pd.concat([top_3_products, bottom_3_products])
ax = sns.barplot(x=selected_products.index, y=selected_products.values, palette='viridis')
for p in ax.patches:
    ax.annotate(format(p.get_height(), '.1f'),
                (p.get_x() + p.get_width() / 2., p.get_height()),
                ha='center', va='center',
                xytext=(0, 9),
                textcoords='offset points')
plt.title('Rating Average per Product Name (Top 5 and Bottom 5)')
plt.xlabel('Product Name')
plt.ylabel('Rating')
plt.xticks(rotation=90)
plt.show()
```
